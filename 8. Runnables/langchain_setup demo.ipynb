## Indivisual approach

```python
import random
```

```python
class LLM:
    def __init__(self):
        print("LLM Created")
        
    def predict(self, prompt:str) -> str:
        response_list = [
            "New delhi is the capital of India.",
            "RCB won the IPL 2025.",
            "Python is a programming language.",
        ]
        return {'response': random.choice(response_list)}
    
llm = LLM()
```

```python
llm.predict("What is the capital of India?")
```

```python
class PromptTemplate:
    def __init__(self, template: str, input_variables:str):
        self.template = template
        self.input_variables = input_variables
        
    def format(self, **kwargs) -> str:
        return self.template.format(**kwargs)
    
template = PromptTemplate(
    template="Write a {length} poem about {country}?",
    input_variables=['length','country']
)

prompt = template.format(**{'length':'short','country':'India'})
```

```python
llm.predict(prompt=prompt)
```

## Chain based approach

```python
class LLMChain:
    def __init__(self,llm, prompt):
        self.llm = llm
        self.prompt = prompt
    def run(self, **kwargs) -> str:
        prompt = self.prompt.format(**kwargs)
        result = self.llm.predict(prompt=prompt)
        return result['response']
    
template = PromptTemplate(
    template="Write a {length} poem about {country}?",
    input_variables=['length','country']
)
llm = LLM()

chain = LLMChain(llm=llm, prompt=template)
chain.run(length='short',country='India')
```
